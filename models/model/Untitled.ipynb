```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from time import sleep

def searchcoordinate(nomecitta):
    driver = webdriver.Firefox()
    driver.get("https://www.gps-longitudine-latitudine.it/converti-indirizzo-coordinate")
    
    # Attendi e gestisci il banner dei cookie
    try:
        # Attendi che il banner dei cookie sia visibile e clicca su "Accetta" o simili
        cookie_button = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, "//button[contains(text(), 'Accetta')]"))
        )
        cookie_button.click()
    except Exception as e:
        print("Banner dei cookie non trovato o non cliccabile:", e)
    
    inputElement = driver.find_element(By.ID, "searchterm")
    inputElement.send_keys(nomecitta)
    #click button 
    driver.find_element(By.CLASS_NAME, 'btn').click()
    #attesa per evitare che i dati siano incompleti 
    sleep(1)
    #elementi che contengono la latitudine e la longitudine 
    lat = driver.find_element(By.ID, "lat").get_attribute("value")
    lng = driver.find_element(By.ID, "lng").get_attribute("value")
    #chiusura webdriver
    driver.close()
    #creazione dizionario con le informazioni utili per effettuare la richiesta alle API meteo  
    infocitta = {'nomecitta': nomecitta, 'latitudine': lat[0:5], 'longitudine': lng[0:5]}
    return infocitta
```
